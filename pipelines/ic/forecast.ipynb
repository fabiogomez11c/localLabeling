```python
from google.cloud import aiplatform
import tensorflow as tf
```

```python
endpoint = aiplatform.Endpoint(
    endpoint_name="projects/813943704943/locations/us-east1/endpoints/3817733070056849408")
```

```python
image = tf.keras.utils.load_img('correct1.png', target_size=(128, 128))
image_array = tf.keras.utils.img_to_array(image)
```

```python
image
```

```python
image_array
```

```python
import numpy as np
image_list = (image_array/255).astype(np.float32).tolist()
```

```python
endpoint.predict(instances=image_list).predictions
```
