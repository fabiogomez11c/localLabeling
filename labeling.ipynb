```python
from pigeon import annotate
from IPython.display import display, Image
import glob
import os
```

```python
# Get all the images

# search in folders and sub-folders
absolute_folder = '/Users/fabio/code/localLabeling/'
data_folder = 'data/splited_data'
image_files = [i.replace(absolute_folder, './') for i in glob.glob(os.path.join(absolute_folder, data_folder)+'/*/*/*.jpg')]

# define the labels
labels = ['front', 'back', 'front_side', 'back_side', 'side']
```

```python
import json

json_files = json.dumps({'filename': image_files})

with open("sample.json", "w") as outfile:
    outfile.write(json_files)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def plot_image(x):
    img = mpimg.imread(x)
    plt.imshow(img)
```
